```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib notebook
```

```python
sns.set(rc={'axes.facecolor':'#242426'}) # graph facecolor
facecolor = '#242426'
font_color = '#ddd'
csfont = {'fontname':'Georgia'} # title font
hfont = {'fontname':'Calibri'} # main font
color_0 = '#fdfe66'
color_1 = '#cdf151'
color_2 = '#88c366'
color_3 = '#008066'
colors = ['#fdfe66', '#88c366', '#cdf151', '#008066', '#888','green', '#ddd', '#fff']
```

```python
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(10, 5), facecolor=facecolor)
fig.tight_layout()

names0 = '1-комн. кв.', '2-комн. кв.', 'студия', '3-комн. кв.', '4-комн. кв.', '5–6-комн. кв.'
sizes0 = [927,577,276,254,41,18]

names1 = 'кирпичный', 'кирп. монолит', 'прочее', 'панельный', 'монолит', 'пан. монолит', 'инд. проект', '137 серия'
sizes1 = [563,526,363,213,149,114,82,70]

_, _, autopcts0 = ax0.pie(sizes0, labels=names0, autopct='%.1f%%', pctdistance=0.75, labeldistance=1.05,
        colors=colors, wedgeprops=dict(width=.5), textprops={'color':font_color})
p0=plt.gcf()

_, _, autopcts1 = ax1.pie(sizes1, labels=names1, autopct='%.1f%%', pctdistance=0.75, labeldistance=1.05,
        colors=colors, wedgeprops=dict(width=.5), textprops={'color':font_color})
p1=plt.gcf()

plt.setp(autopcts0, **{'color':'#333', 'weight':'bold', 'fontsize':12.5})
plt.setp(autopcts1, **{'color':'#333', 'weight':'bold', 'fontsize':12.5})

title = fig.suptitle('Популярные категории арендного жилья в Санкт-Петербурге', 
                     y=.95, fontsize=18, color=font_color, **csfont)
# To prevent the title from being cropped
plt.subplots_adjust(top=0.85, bottom=None, right=0.9)
```
