```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
file = 'data/spb-rentals-2020-feb'
df = pd.read_csv(file + '.csv', encoding='utf-16', delimiter='\t')
```

```python
df['Аренда, руб./мес.'] = df['Аренда, руб./мес.'] / 1000
df.head()
```

```python
x0 = df[df['Тип жилья'] == 'студия']['Аренда, руб./мес.']
x1 = df[df['Тип жилья'] == '1-комн. квартира']['Аренда, руб./мес.']
x2 = df[df['Тип жилья'] == '2-комн. квартира']['Аренда, руб./мес.']
x3 = df[df['Тип жилья'] == '3-комн. квартира']['Аренда, руб./мес.']
x4 = df[df['Тип жилья'] == '4-комн. квартира']['Аренда, руб./мес.']
x5 = df[df['Тип жилья'] == '5-комн. квартира']['Аренда, руб./мес.']
x6 = df[df['Тип жилья'] == '6-комн. квартира']['Аренда, руб./мес.']
```

```python
sns.set(rc={'axes.facecolor':'#242426'}) # graph facecolor
facecolor = '#242426'
font_color = '#ddd'
csfont = {'fontname':'Georgia'} # title font
hfont = {'fontname':'Calibri'} # main font
color_0 = '#fdfe66'
color_1 = '#cdf151'
color_2 = '#88c366'
color_3 = '#008066'
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8), sharey=True, dpi=100, facecolor=facecolor)
fig.tight_layout()

ax0 = sns.histplot(data=x0, x=x0.values, ax=axes[0, 0], color=color_0, zorder=3, binwidth=3)
ax1 = sns.histplot(data=x1, x=x1.values, ax=axes[0, 1], color=color_1, zorder=3, binwidth=6)
ax2 = sns.histplot(data=x2, x=x2.values, ax=axes[1, 0], color=color_2, zorder=3, binwidth=9)
ax3 = sns.histplot(data=x3, x=x3.values, ax=axes[1, 1], color=color_3, zorder=3)

mpl.rcParams['axes.linewidth'] = .5
ax0.grid(color=font_color, linestyle='-', linewidth=.5, zorder=0)
ax1.grid(color=font_color, linestyle='-', linewidth=.5, zorder=0)
ax2.grid(color=font_color, linestyle='-', linewidth=.5, zorder=0)
ax3.grid(color=font_color, linestyle='-', linewidth=.5, zorder=0)

ax0.tick_params(colors=font_color, grid_color=font_color)
ax1.tick_params(colors=font_color, grid_color=font_color)
ax2.tick_params(colors=font_color, grid_color=font_color)
ax3.tick_params(colors=font_color, grid_color=font_color)
    
ax0.set_ylabel('Количество', color=font_color, fontsize=15, **hfont)
ax2.set_ylabel('Количество', color=font_color, fontsize=15, **hfont)

ax0.set_xlabel('Аренда студий, тыс. руб./мес.', color=font_color, fontsize=15, **hfont)
ax1.set_xlabel('Аренда 1-комн. квартир, тыс. руб./мес.', color=font_color, fontsize=15, **hfont)
ax2.set_xlabel('Аренда 2-комн. квартир, тыс. руб./мес.', color=font_color, fontsize=15, **hfont)
ax3.set_xlabel('Аренда 3-комн. квартир, тыс. руб./мес.', color=font_color, fontsize=15, **hfont)

ax2.set_xlim(0,180)

title = 'Количество предложений по ценовым категориям и типам квартир'
fig.suptitle(title, y=.92, fontsize=18, color=font_color, **csfont)

plt.subplots_adjust(left=None, bottom=0.3, right=None, top=0.85, wspace=None, hspace=.6)
```
