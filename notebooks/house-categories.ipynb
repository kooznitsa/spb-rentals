```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib notebook
```

```python
file = 'data/spb-rentals-2020-feb'
df = pd.read_csv(file + '.csv', encoding='utf-16', delimiter='\t')
df = df.drop_duplicates()
df.head()
```

```python
df['Этаж'] = df['Этаж'].str.split().str.get(0).str.split('/').str.get(1)
df.head()
```

```python
houses = df.groupby(['Тип дома'])[['Год постройки', 'Этаж', 'Станция метро', 'Общая площадь, м']].agg(lambda x: list(x.mode()))
houses
```
