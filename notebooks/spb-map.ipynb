```python
file = 'data/spb-rentals-2020-feb'
df = pd.read_csv(file + '.csv', encoding='utf-16', delimiter='\t')
df.head()
```

```python
df['Аренда, руб./мес.'] = df['Аренда, руб./мес.'] / df['Общая площадь, м']
df
```

```python
df1 = df[['X', 'Y', 'Аренда, руб./мес.']]
df1 = df1.rename({'Аренда, руб./мес.':'Rent'}, axis='columns')
df1
```

```python
df1.to_csv('data/coordinates-2020-feb.csv')
```

```python
data = gmaps.datasets.load_dataset_as_df('spb_rent')
data.head()
```

```python
figure_layout = {
    'width': '100%',
    'height': '75vh'
}

locations = data[['X', 'Y']]
weights = data['Rent']
coordinates = (59.9311, 30.3609)

fig = gmaps.figure(center=coordinates, 
                   zoom_level=10, 
                   map_type='HYBRID', 
                   layout=figure_layout)

fig.add_layer(gmaps.heatmap_layer(locations, 
                                  weights=weights, 
                                  point_radius=5, 
                                  max_intensity=3000))
fig
```

```python
embed_minimal_html('spb-rent-map.html', views=[fig])
```
